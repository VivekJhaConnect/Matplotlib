```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
goog_df =  pd.read_csv(r'..\datasets\Yahoo-Finance-GOOG.csv')
goog_df['Open'] = pd.Series([round(val, 2) for val in goog_df['Open']], index=goog_df.index)
goog_df['High'] = pd.Series([round(val, 2) for val in goog_df['High']], index=goog_df.index)
goog_df['Low'] = pd.Series([round(val, 2) for val in goog_df['Low']], index=goog_df.index)
goog_df['Close'] = pd.Series([round(val, 2) for val in goog_df['Close']], index=goog_df.index)
goog_df['Adj Close'] = pd.Series([round(val, 2) for val in goog_df['Adj Close']], index=goog_df.index)

goog_df.head()
```

```python
stk_data = goog_df[-5:]
col_head = ('Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume')
stk_data_np = stk_data.to_numpy()

plt.figure(linewidth=2, figsize=(5, 3))
axes_2 = plt.gca()
axes_2.get_xaxis().set_visible(False)
axes_2.get_yaxis().set_visible(False)
plt.box(on=None)
plt.tight_layout(pad=0.5)  # Corrected 'Pad' to 'pad'

colors = plt.cm.Blues(np.full(len(col_head), 0.2))

the_table = plt.table(cellText=stk_data_np, loc='center', colLabels=col_head,
                      colColours=colors)

the_table.set_fontsize(14)
the_table.scale(3, 2.5)
```
