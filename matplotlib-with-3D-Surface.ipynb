```python
fig_1 = plt.figure(figsize=(8,5), dpi=100)
axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9], projection='3d')


def get_z(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x_2 = np.linspace(-6, 6, 30)
y_2 = np.linspace(-6, 6, 30)

x_2, y_2 = np.meshgrid(x_2, y_2)

z_2 = get_z(x_2, y_2)

axes_1.view_init(45, 55)
axes_1.contour3D(x_2, y_2, z_2, 80, cmap='Blues')
```
