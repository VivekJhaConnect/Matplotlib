```python
ics_df = pd.read_csv('..\datasets\Icecream.csv', sep=",")
ics_df.head()

ics_df_2 =ics_df[['temp','income']]
ics_df_2 = ics_df_2.sort_values(by='temp')
ics_df_2.head()
```

```python
np_arr = ics_df_2.values
x_2 = np_arr[:, 0]
y_2 = np_arr[:, 1]

fig_1 = plt.figure(figsize=(6,4))
axes_1 = fig_1.add_axes([0,0,1,1])

axes_1.set_xlabel('Temp')  # Corrected method name
axes_1.set_ylabel('Ice Cream Sales')
axes_1.set_title('Ice Cream Sales vs. Temp')
axes_1.plot(x_2, y_2)

axes_1.annotate('Good Month', xy=(20, 90), xytext=(20, 90), arrowprops=dict(facecolor='black', shrink=0.05))

plt.bar(x_2, y_2)
```
