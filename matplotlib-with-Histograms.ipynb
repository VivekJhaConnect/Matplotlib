```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
arr_1 = np.random.randint(1,7,7000)
arr_2 = np.random.randint(1,7,7000)

arr_3 = arr_1+ arr_2
plt.hist(arr_3, bins=11, density=True, stacked=True)

# Range gets tuple with bin range interested in 
# cumulative=True use a cumulative distribution 
# histtype='step' genrates a line plot
# orientation='horizontal'
# color='orange' change bar color
```

```python
plt.hist(arr_3, bins=11, density=True, stacked=True, cumulative=True)
```

```python
plt.hist(arr_3, bins=11, density=True, stacked=True,  histtype='step')
```

```python
plt.hist(arr_3, bins=11, density=True, stacked=True,  orientation='horizontal')
```

```python
plt.hist(arr_3, bins=11, density=True, stacked=True,  orientation='horizontal', color='orange')
```
