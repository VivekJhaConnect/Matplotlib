```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
fig_1 = plt.figure(figsize=(8,5), dpi=100)
axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9])

# Create a pie chart of the number of Pokemon by tyep

tyeps = ['Water', 'Normal', 'Flying', 'Grass', 'Psychic', 'Bug', 'Fire', 'Poison',
         'Ground', 'Rock', 'Fighting', 'Dark', 'Steel', 'Electric', 'Dragon', 'Fairy',
         'Ghost', 'Ice']

poke_num = [133,109,101,98,77,68,65,60,57,54,53,51,50,50,46, 40]

colours = []

for i in range(18):
    rgb = (random.uniform(0,0.5), random.uniform(0, 0.5), random.uniform(0, 0.5))
    colours.append(rgb)

explode = [0] * 18
explode[0] = 0.2

wedges, texts, autotexts = plt.pie(poke_num, explode=explode, labels=tyeps,
                                   colors=colours, autopct='%1.0f%%', shadow=True,
                                   startangle=140, textprops=dict(colour='w'))
plt.legend(wedges, tyeps, loc='right', bbox_to_anchor=(1,0,0.5,1))
```
